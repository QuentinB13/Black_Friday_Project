# Black Friday Project

Import Black Friday Dataset

```python
import pandas as pd

bftrain = pd.read_csv('train.csv')
bftest = pd.read_csv('test.csv')
```

```python
bftrain
```

```python
bftest['Purchase'] = 0
bftest
```

```python
df = pd.concat([bftrain, bftest], axis = 0)
df
```

Split X & Y: The point here is to combine train and test to apply all the feature engineering to both at the same time.

```python
target = df.Purchase
df.drop(['Purchase'], axis = 1, inplace = True)
```

Data Cleaning: It consists of the transformation of all categorical data and all numerical data.

```python
df.dtypes
```

```python
import numpy as np

df = df.applymap(str)
df.dtypes
```

Categorical Data: All the categories are transformed.

```python
for i in ['User_ID', 'Product_ID', 'Gender', 'Occupation', 'City_Category', 'Marital_Status']:
    df[str(i)] = df[str(i)].astype('category')

df.dtypes
```

Numerical Data: All the numbers are transformed.

```python
df.Age.unique()
```

```python
df.Age.replace(['0-17'], ['15'], inplace = True)
df.Age.replace(['18-25'], ['21'], inplace = True)
df.Age.replace(['26-35'], ['30'], inplace = True)
df.Age.replace(['36-45'], ['41'], inplace = True)
df.Age.replace(['46-50'], ['48'], inplace = True)
df.Age.replace(['51-55'], ['53'], inplace = True)
df.Age.replace(['55+'], ['60'], inplace = True)

df.Age.unique()
```

```python
df.Stay_In_Current_City_Years.unique()
```

```python
df.Stay_In_Current_City_Years.replace(['4+'], ['4'], inplace = True)

df.Stay_In_Current_City_Years.unique()
```

```python
df.Product_Category_1.unique()
```

```python
df.Product_Category_2.unique()
```

```python
df.Product_Category_2.replace(['nan'], ['-1'], inplace = True)
df.Product_Category_2.unique()
```

```python
df.Product_Category_3.unique()
```

```python
df.Product_Category_3.replace(['nan'], ['-1'], inplace = True)
df.Product_Category_3.unique()
```

```python
for i in ['Age', 'Stay_In_Current_City_Years', 'Product_Category_1', 'Product_Category_2', 'Product_Category_3']:
    df[str(i)] = df[str(i)].astype('float')

df.dtypes
```

```python
df.Product_Category_2.replace([-1], [round(df.Product_Category_2.mean())], inplace = True)
df.Product_Category_2.unique()
```

```python
df.Product_Category_3.replace([-1], [round(df.Product_Category_3.mean())], inplace = True)
df.Product_Category_3.unique()
```

NA Management: The point here is to find a way for every variable to replace NAs by relevant values.

```python
df.isna().sum()
```

Dummification: The point here is to dummify the categorical data.

```python
dummies = pd.get_dummies(df[['Gender', 'City_Category', 'Marital_Status']])
dummies
```

```python
df = pd.concat([df, dummies], axis = 1)

df
```

```python
df = df.drop(columns = ['Gender', 'City_Category', 'Marital_Status'])

df
```

SPLIT TRAIN TEST

```python
full = pd.concat([df, target], axis = 1)
full.dtypes
```

```python
full2 = full.drop(columns = ['User_ID', 'Product_ID', 'Occupation'])
```

```python
train = full2.loc[full2.Purchase != 0,:]
test = full2.loc[full2.Purchase == 0,:]
```

```python
x_train = train.drop(columns = ['Purchase'])
y_train = train.Purchase

x_test = test.drop(columns = ['Purchase'])
```

MODEL 1: XGBOOST REGRESSOR

```python
import xgboost

model = xgboost.XGBRegressor(colsample_bytree=0.4,
                 gamma=0,                 
                 learning_rate=0.07,
                 max_depth=3,
                 min_child_weight=1.5,
                 n_estimators=10000,                                                                    
                 reg_alpha=0.75,
                 reg_lambda=0.45,
                 subsample=0.6,
                 seed=42)
```

```python
x_train.dtypes
```

```python
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
predictions = pd.DataFrame(predictions)
predictions

solution = pd.concat([bftest[['User_ID', 'Product_ID']], predictions], axis = 1)
solution.columns = ['User_ID', 'Product_ID', 'Purchase'
```

```python
solution.to_csv('Solution1.csv', index = False)
```

TEST ON TRAIN BEFORE FINAL TEST

```python
from sklearn.model_selection import train_test_split
np.random.seed(123)
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size = 0.2)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
```
